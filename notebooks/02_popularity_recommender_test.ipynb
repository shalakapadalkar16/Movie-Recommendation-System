```python
# testing the window based trending recommender

import sys
from pathlib import Path

PROJECT_ROOT = Path().resolve().parent
sys.path.append(str(PROJECT_ROOT))
```

```python
from src.data.load_data import load_ratings, load_movies
from src.recommenders.popularity_recommender import TrendingRecommender
```

```python
DATA_PATH = PROJECT_ROOT / "data" / "raw"

ratings = load_ratings(DATA_PATH)
movies = load_movies(DATA_PATH)

print(ratings.shape)
print(movies.shape)
```

```python
trending_rec = TrendingRecommender(window_days=365)
trending_rec.fit(ratings, movies)

trending_rec.recommend(top_k=10)
```

```python
for days in [30, 90, 180, 365]:
    trending_rec = TrendingRecommender(window_days=days)
    trending_rec.fit(ratings, movies)
    print(f"\nTop 5 trending movies in last {days} days:")
    display(trending_rec.recommend(top_k=5))
```

```python
from src.recommenders.popularity_recommender import TrendingWeightedRecommender

tw_rec = TrendingWeightedRecommender(window_days=90, min_votes=50)
tw_rec.fit(ratings, movies)

tw_rec.recommend(10)
```
